# Copper Concentrate Market Overview

This notebook provides a short companion overview for the Streamlit dashboard. It uses illustrative sample data rather than live market feeds, which keeps the project reproducible and avoids paid market-data dependencies.

Copper concentrate valuation combines market price exposure with physical quality, payable terms, smelter charges, logistics, and commercial deductions.

## Main Commercial Drivers

- Copper price in USD per metric tonne
- Dry metric tonnes after moisture adjustment
- Copper grade and payable copper percentage
- Treatment charge in USD/dmt
- Refining charge in US cents/lb of payable copper
- Freight and logistics cost
- Impurity penalties
- Gold and silver by-product credits

```python
import pandas as pd
import matplotlib.pyplot as plt

prices = pd.read_csv('../data/sample_lme_prices.csv')
specs = pd.read_csv('../data/sample_concentrate_specs.csv')

prices.head(), specs.head()
```

## Illustrative Copper Price Series

The dashboard can use this sample series to show how one selected cargo valuation changes with indicative market assumptions. Cargo quality is held constant; copper price, TC/RC, freight, precious metal prices, FX, and financing rate vary by date. This is not live LME data.

```python
prices['date'] = pd.to_datetime(prices['date'])
prices.plot(x='date', y='copper_price_usd_per_tonne', marker='o', legend=False)
plt.ylabel('USD per tonne')
plt.title('Illustrative Copper Price Series')
plt.show()
```

## Sample Concentrate Comparison

A useful commercial habit is comparing cargo quality and terms side by side before jumping to value. Higher grade is helpful, but freight, penalties, payable terms, and by-product credits can change the ranking.

```python
comparison_columns = [
    'concentrate_name',
    'origin_region',
    'copper_grade_percentage',
    'moisture_percentage',
    'tc_usd_per_dmt',
    'rc_cents_per_lb',
    'freight_usd_per_dmt',
    'impurity_penalty_usd_per_dmt',
]
specs[comparison_columns].sort_values('copper_grade_percentage', ascending=False)
```

## Note on TC and RC Units

TC is charged per dry metric tonne of concentrate because it relates to treating the bulk material. RC is charged per pound of payable copper because it relates to refining the payable metal units. This is why the dashboard explicitly converts payable copper tonnes into pounds.
