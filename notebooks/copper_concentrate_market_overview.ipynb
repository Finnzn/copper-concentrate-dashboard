## Main Commercial Drivers

- Copper price in USD per metric tonne
- Dry metric tonnes after moisture adjustment
- Copper grade and payable copper percentage
- Treatment charge in USD/dmt
- Refining charge in US cents/lb of payable copper
- Freight and logistics cost
- Impurity penalties
- Gold and silver by-product credits

## Note on TC and RC Units

TC is charged per dry metric tonne of concentrate because it relates to treating the bulk material. RC is charged per pound of payable copper because it relates to refining the payable metal units. This is why the dashboard explicitly converts payable copper tonnes into pounds.
